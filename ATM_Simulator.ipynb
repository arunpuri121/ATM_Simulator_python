```python
class atm_simulator(): # this is the class for atm simulator 
    
    def __init__(self): # contstructor 
        self.balance = 10000
        self.pin = 1234
    
    def check_balance(self): # check balance method
        print(f"\nYour available balance is {self.balance}\n")
    
    def withdrawls(self, withdrawl): # Withdrawl method
        self.balance -= withdrawl
        print(f"\nYou have withdrawn: {withdrawl} rupees")
        print(f"\nYour available balance is {self.balance}\n")
        
    def deposit(self, deposit): # deposit method
        self.balance += deposit
        print(f"\nYou have deposited: {deposit} rupees")
        print(f"\nYour available balance is {self.balance}\n")

        
def atm_sim(atm): #function for calculating different outputs
    
    print("*********** Welcome To The ATM ************\n")
    pin_check = int(input("Please Enter Your PIN: ")) # taking input for pin
    
    if atm.pin == pin_check: # condition check for coreect pin
        print("\n*******Pin is Correct*******")
        print("\nPlease select One of the Options")
        print("\n1. Balance Enquiry, 2. Withdrawls, 3. Deposits, 4. Exit\n")
        option = int(input("Please type a number(i.e - 1. for balace enquiry, 2. for Withdrawls and so on...) "))
        
        if option == 1: # for displaying current balance
            atm.check_balance()
            print(f"Do you want to do another transaction ?\n\nPress 1 for Yes or Press any number for NO\n")
            cont = int(input(""))
            if cont == 1:
                atm_sim(atm)
            else:
                print("\nThank you For Using ATM Simulator")
            
        elif option ==2: # for withdrawl of money
            withdrawl = int(input("\nPlease enter the amount to be withdrawn:"))
            if atm.balance>=withdrawl:
                atm.withdrawls(withdrawl)
                
                print(f"Do you want to do another transaction ?\n\nPress 1 for Yes or Press any number for NO")
                cont = int(input(""))
                if cont == 1:
                    atm_sim(atm)
                else:
                    print("\nThank you For Using ATM Simulator")
                
            else:
                print("\ninsufficient balance")
                atm.check_balance()
                
                print(f"Do you want to do another transaction ?\n\nPress 1 for Yes or Press any number for NO")
                cont = int(input(""))
                if cont == 1:
                    atm_sim(atm)
                else:
                    print("\nThank you For Using ATM Simulator")
        
        elif option == 3: # for deposit of money
            deposit = int(input("\nPlease enter the amount to deposit: "))
            atm.deposit(deposit)
                
            print(f"Do you want to do another transaction ?\n\nPress 1 for Yes or Press any number for NO")
            cont = int(input(""))
            if cont == 1:
                atm_sim(atm)
            else:
                print("\nThank you For Using ATM Simulator")
        
        elif option == 4: # for exit
            print("\nYou have been exited. Thank You For Using ATM")
    else:
        print("\nIncorrect pin Transaction Cancelled!!")
        
if __name__=="__main__":
    atm = atm_simulator() # caliing the constructor and making the object atm
    atm_sim(atm) # calling the function to run the codes
```
